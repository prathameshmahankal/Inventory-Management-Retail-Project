```python
df_etc = df[(df['Store No.']==111300) & (df['Day']!='Saturday')]
```

## Working with the 15-min counts per day for the quarter ##

Next, let's group the number of transactions according to the day of the week and the 15 minute time bucket of the transactions.

```python
df_grouped = df_etc[['Day','Time','No. of Transactions']].groupby(['Time','Day']).sum()
df_grouped.head()
```

```python
df_grouped = df_grouped.unstack()
df_grouped.head()
```

```python
df_grouped.columns
```

```python
fig, ax = plt.subplots(figsize=(14, 8), dpi=80, facecolor='w', edgecolor='k')
df_grouped.plot(kind='line', ax=ax)
ax.legend(['Friday', 'Monday', 'Thursday', 'Tuesday', 'Wednesday']);
ax.grid(True, color="#93a1a1", alpha=0.3)
ax.set_xlabel('15-min intervals per day', fontsize=15)
ax.set_ylabel('Number of transactions', fontsize=15)
ax.tick_params(labelrotation=90)
```

```python
df_grouped1 = df_etc[['Day','Time','Date','No. of Transactions']].groupby(['Day','Time','Date']).sum()
df_grouped1 = df_grouped1.unstack()
df_grouped1 = df_grouped1.fillna(0)
df_grouped1.head()
```

```python
df_grouped1.reset_index().columns
```

```python
pd.melt(df_grouped1.loc[(slice('Friday'), slice(None)),])
```

```python
df_grouped1['Mean'] = df_grouped1['No. of Transactions'].apply(lambda x:np.mean(x), axis=1)
df_grouped1['StdDev'] = df_grouped1['No. of Transactions'].apply(lambda x:np.std(x), axis=1)
df_grouped1['Variance'] = df_grouped1['No. of Transactions'].apply(lambda x:np.var(x), axis=1)
```

```python
df_grouped1[['Mean', 'StdDev', 'Variance']].head(10)
```
